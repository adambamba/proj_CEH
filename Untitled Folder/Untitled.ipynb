```python
import matplotlib.pyplot as plt
import xarray as xr
from utils import u_plot
import numpy as np
from utils import u_interpolate as uint
```

```python
ds=xr.open_dataset('/users/global/cornkle/test_irregular.nc')
sh = ds['SH']
shv = sh.values[10,:,:]
```

```python
u_plot.quick_map(shv, sh['longitude_t'], sh['latitude_t'], cmap='jet', levels=[-20, -10, -5, 5, 10 ,20])
```

```python
lats = sh.latitude_t[:, 0].values
lons = sh.longitude_t[0, :].values
latmin = sh.latitude_t.min()
latmax = sh.latitude_t.max()
lonmin = sh.longitude_t.min()
lonmax = sh.longitude_t.max()
dist = np.mean((lats)[1::] - (lats)[0:-1])

lat_regular = np.arange(latmin+10*dist, latmax-10*dist, dist)
lon_regular = np.arange(lonmin+10*dist, lonmax-10*dist, dist)
```

```python
regridded1 = uint.griddata_comparison(shv, sh.longitude_t.values, sh.latitude_t.values, lon_regular, lat_regular, isll=True)
```

```python
u_plot.quick_map(regridded1, lon_regular, lat_regular, cmap='jet', levels=[-20, -10, -5, 5, 10 ,20])
```

```python
regridded2 = uint.griddata_comparison(shv, sh.longitude_t.values, sh.latitude_t.values, lon_regular, lat_regular, isll=False)
```

```python
u_plot.quick_map(regridded2, lon_regular, lat_regular, cmap='jet', levels=[-20, -10, -5, 5, 10 ,20])
```

```python
u_plot.quick_map(regridded1-regridded2, lon_regular, lat_regular, cmap='RdBu', extent='both', levels=[-0.5,-0.2, -0.1, 0.1,0.2, 0.5])
```

```python
u_plot.quick_map(regridded1-regridded2, lon_regular, lat_regular, cmap='RdBu', extent='both', levels=[-0.5,-0.2, -0.1, 0.1,0.2, 0.5])
```
