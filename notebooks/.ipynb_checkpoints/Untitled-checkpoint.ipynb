```python
outt=dic['tc_lag0'].values.copy()
mmeans=np.percentile(outt[np.isfinite(outt)], 30)
print(mmeans)
```

```python
outp=dic['p'].values.copy()
outp[np.isnan(outp)]=-10**-5
outt[np.isnan(outt)]=-150  
#mmeans=dic.meanT_cut
```

```python
grad=np.gradient(outt)
plt.pcolormesh(grad[1])
plt.colorbar()

nok = np.where(abs(grad[1]) > 40)
```

```python
o2=outt.copy()
#o1=outt.copy()
o2[o2<-100]=mmeans
o2[nok[::2,]]=mmeans
o2[nok[1::2]]=-50
d=2
i=nok[0]
j=nok[1]
```

```python
for ii,jj in zip(i,j):
    
    kernel=o2[ii-d:ii+d+1, jj-d:jj+d+1]
    if not kernel.any():
        o2[ii,jj]=mmeans
    else:    
        o2[ii-d:ii+d+1, jj-d:jj+d+1]=ndimage.gaussian_filter(kernel, 3)
```

```python
plt.pcolormesh(o2)
plt.colorbar()
plt.title('Smooth edges')
```

```python
pos=10
wav = util.waveletTP(o2, outp, 5)
print(wav['scales'][pos])
tt=wav['t'][pos,:,:]
pp=wav['p'][pos,:, :]
tt[np.where(dic['pmask'].values==0)]=0
#pm=pp.max()
#tm=tt.max()
#tt=tt/tm
#pp=pp/pm
#tt[tt>0.1]=1
#pp[pp>0.1]=1
#tt[tt<=0.1]=0
#pp[pp<=0.1]=0
```
