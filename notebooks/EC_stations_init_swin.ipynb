```python
# common way to import is
import pandas as pd
from pandas import *
import math
import numpy as np
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
#import seaborn as sns
%matplotlib inline
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = (16., 9.)
#sns.set_palette("deep", desat=.6)
#sns.set_context(rc={"figure.figsize": (8, 4)})

# we need this to parse the uncommon timestamp
from datetime import datetime
```

```python
def join_ec_data():
    # Read the CSV data
    #RADIATION FILES
    df1 = pd.read_table('/media/klein-c/Elements/Obs_Data/EC_DATA/radiation/wascal-1_Radiation.csv',delimiter=',', parse_dates=['Date'])
    df2 = pd.read_table('/media/klein-c/Elements/Obs_Data/EC_DATA/radiation/wascal-2_Radiation.csv',parse_dates=['Date'])
    df3 = pd.read_table('/media/klein-c/Elements/Obs_Data/EC_DATA/radiation/wascal-3_Radiation.csv',delimiter=',', parse_dates=['Date'])
    df1.set_index('Date', inplace=True, drop=True)
    df2.set_index('Date', inplace=True, drop=True)
    df3.set_index('Date', inplace=True, drop=True)
    
    df1.SR_in_cor[df1.SR_in_cor<-0]=0
    df1.SR_out_cor[df1.SR_in_cor==0]=0
    df2.SR_in_cor[df2.SR_in_cor<-0]=0
    df2.SR_out_cor[df2.SR_in_cor==0]=0
    df3.SR_in_cor[df3.SR_in_cor<-0]=0
    df3.SR_out_cor[df3.SR_in_cor==0]=0
    df1.IR_out_corr=df1.IR_out_corr+df1.CNR4TK_Avg**4*5.6698e-8
    df2.IR_out_corr=df2.IR_out_corr+df2.CNR4TK_Avg**4*5.6698e-8
    df3.IR_out_corr=df3.IR_out_corr+df3.CNR4TK_Avg**4*5.6698e-8
    df1.IR_in_corr=df1.IR_in_corr+df1.CNR4TK_Avg**4*5.6698e-8
    df2.IR_in_corr=df2.IR_in_corr+df2.CNR4TK_Avg**4*5.6698e-8
    df3.IR_in_corr=df3.IR_in_corr+df3.CNR4TK_Avg**4*5.6698e-8
    
    df1_g=df1.loc[:, ['IR_in_corr', 'IR_out_corr', 'SR_in_cor', 'SR_out_cor']].resample('H')
    df2_g=df2.loc[:,['IR_in_corr', 'IR_out_corr', 'SR_in_cor', 'SR_out_cor']].resample('H')
    df3_g=df3.loc[:,['IR_in_corr', 'IR_out_corr', 'SR_in_cor', 'SR_out_cor']].resample('H')
    
    df1_g.columns=['EW1_LW_in', 'EW1_LW_out', 'EW1_SW_in', 'EW1_SW_out']
    df2_g.columns=['EW2_LW_in', 'EW2_LW_out', 'EW2_SW_in', 'EW2_SW_out']
    df3_g.columns=['EW3_LW_in', 'EW3_LW_out', 'EW3_SW_in', 'EW3_SW_out']
   
    
    # Read the CSV data
    # TK3 FLUX FILES
    df4 = pd.read_table('/media/klein-c/Elements/Obs_Data/EC_DATA/radiation/wascal-1_TK3_output.csv',delimiter=',', parse_dates=['Date'])
    df5 = pd.read_table('/media/klein-c/Elements/Obs_Data/EC_DATA/radiation/wascal-2_TK3_output.csv',delimiter=',',parse_dates=['Date'])
    df6 = pd.read_table('/media/klein-c/Elements/Obs_Data/EC_DATA/radiation/wascal-3_TK3_output.csv',delimiter=',', parse_dates=['Date'])
    df4.set_index('Date', inplace=True, drop=True)
    df5.set_index('Date', inplace=True, drop=True)
    df6.set_index('Date', inplace=True, drop=True)
    df4_g=df4.loc[:, ['LvE[W/m²]', 'HTs[W/m²]', 'Ts[°C]', 'T_ref[°C]']].resample('H')
    df5_g=df5.loc[:, ['LvE[W/m²]', 'HTs[W/m²]', 'Ts[°C]', 'T_ref[°C]']].resample('H')
    df6_g=df6.loc[:, ['LvE[W/m²]', 'HTs[W/m²]', 'Ts[°C]', 'T_ref[°C]']].resample('H')
    df4_g.columns=['EW1_LFX', 'EW1_HFX', 'EW1_T', 'EW1_Tref']
    df5_g.columns=['EW2_LFX', 'EW2_HFX', 'EW2_T', 'EW2_Tref']
    df6_g.columns=['EW3_LFX', 'EW3_HFX', 'EW3_T', 'EW3_Tref']
    
    # Read the CSV data
    # SOIL FILES
    df7 = pd.read_table('/media/klein-c/Elements/Obs_Data/EC_DATA/radiation/wascal-1_Soil_variables.csv', parse_dates=['Time'])
    df8 = pd.read_table('/media/klein-c/Elements/Obs_Data/EC_DATA/radiation/wascal-2_Soil_variables.csv',parse_dates=['Time'])
    df9 = pd.read_table('/media/klein-c/Elements/Obs_Data/EC_DATA/radiation/wascal-3_Soil_variables.csv', parse_dates=['Time'])
    df7.set_index('Time', inplace=True, drop=True)
    df8.set_index('Time', inplace=True, drop=True)
    df9.set_index('Time', inplace=True, drop=True)
    df7_g=df7.loc[:, ['VWC_3cm_Avg', 'SHF_Avg']].resample('H')
    df8_g=df8.loc[:,['VWC_3cm_Avg', 'SHF_Avg']].resample('H')
    df9_g=df9.loc[:,['VWC_3cm_Avg', 'SHF_Avg']].resample('H')
    df7_g.columns=['EW1_VWC', 'EW1_GFX']
    df8_g.columns=['EW2_VWC', 'EW2_GFX']
    df9_g.columns=['EW3_VWC', 'EW3_GFX']
    df7_g.EW1_GFX[df7_g.EW1_GFX<-300]=np.NaN
    df8_g.EW2_GFX[df8_g.EW2_GFX<-300]=np.NaN
    df9_g.EW3_GFX[df9_g.EW3_GFX<-300]=np.NaN
    
    # Read Rainfall CSV
    df10 = pd.read_csv('/media/klein-c/Elements/Obs_Data/EC_DATA/radiation/wascal-1_Cr1000.csv', parse_dates=['TIMESTAMP']).dropna(how='any')
    df11 = pd.read_csv('/media/klein-c/Elements/Obs_Data/EC_DATA/radiation/wascal-2_Cr1000.csv', parse_dates=['TIMESTAMP']).dropna(how='any')
    df12 = pd.read_csv('/media/klein-c/Elements/Obs_Data/EC_DATA/radiation/wascal-3_Cr1000.csv', parse_dates=['TIMESTAMP']).dropna(how='any')
    df10.set_index('TIMESTAMP', inplace=True, drop=True)
    df11.set_index('TIMESTAMP', inplace=True, drop=True)
    df12.set_index('TIMESTAMP', inplace=True, drop=True)
    df10_g=df10.loc[:,['Rain_mm_Tot']].resample('H', how='sum')
    df11_g=df11.loc[:,['Rain_mm_Tot']].resample('H', how='sum')
    df12_g=df12.loc[:,['Rain_mm_Tot']].resample('H', how='sum')
    df10_g.columns=['EW1_PRCP']
    df11_g.columns=['EW2_PRCP']
    df12_g.columns=['EW3_PRCP']
    
        # Read Rainfall CSV
    df13 = pd.read_csv('/media/klein-c/Elements/Obs_Data/EC_DATA/radiation/wascal-1_WXT.csv', parse_dates=['TIMESTAMP']).dropna(how='any')
    df14 = pd.read_csv('/media/klein-c/Elements/Obs_Data/EC_DATA/radiation/wascal-2_WXT.csv', parse_dates=['TIMESTAMP']).dropna(how='any')
    df15 = pd.read_csv('/media/klein-c/Elements/Obs_Data/EC_DATA/radiation/wascal-3_WXT.csv', parse_dates=['TIMESTAMP']).dropna(how='any')
    df13.set_index('TIMESTAMP', inplace=True, drop=True)
    df14.set_index('TIMESTAMP', inplace=True, drop=True)
    df15.set_index('TIMESTAMP', inplace=True, drop=True)
    df13_g=df13.loc[:,['Wsavg_Avg']].resample('H', how='mean')
    df14_g=df14.loc[:,['Wsavg_Avg']].resample('H', how='mean')
    df15_g=df15.loc[:,['Wsavg_Avg']].resample('H', how='mean')
    df13_g.columns=['EW1_WS']
    df14_g.columns=['EW2_WS']
    df15_g.columns=['EW3_WS']
    
    # JOIN FILES
    df_join=df1_g.join([df2_g, df3_g, df4_g, df5_g, df6_g, df7_g, df8_g, df9_g, df10_g, df11_g, df12_g, df13_g, df14_g, df15_g], how='outer')
    
    # COMPUTE AND ADD NET RADIATION AND FLUX SUMS
    df_join['EW1_rad_net']=df_join.EW1_SW_in-df_join.EW1_SW_out+df_join.EW1_LW_in-df_join.EW1_LW_out
    df_join['EW2_rad_net']=df_join.EW2_SW_in-df_join.EW2_SW_out+df_join.EW2_LW_in-df_join.EW2_LW_out
    df_join['EW3_rad_net']=df_join.EW3_SW_in-df_join.EW3_SW_out+df_join.EW3_LW_in-df_join.EW3_LW_out
    
    df_join['EW1_rad_in']=df_join.EW1_SW_in+df_join.EW1_LW_in
    df_join['EW2_rad_in']=df_join.EW2_SW_in+df_join.EW2_LW_in
    df_join['EW3_rad_in']=df_join.EW3_SW_in+df_join.EW3_LW_in
    
    df_join['EW1_flux_sum']=df_join.EW1_LFX+df_join.EW1_HFX +df_join.EW1_GFX
    df_join['EW2_flux_sum']=df_join.EW2_LFX+df_join.EW2_HFX +df_join.EW2_GFX
    df_join['EW3_flux_sum']=df_join.EW3_LFX+df_join.EW3_HFX +df_join.EW3_GFX
    
    df_join['EW1_Residual']=df_join.EW1_rad_net-df_join.EW1_flux_sum
    df_join['EW2_Residual']=df_join.EW2_rad_net-df_join.EW2_flux_sum
    df_join['EW3_Residual']=df_join.EW3_rad_net-df_join.EW3_flux_sum
    
    df_join['EW1_ALBEDO']=(df_join.EW1_SW_out/df_join.EW1_SW_in)*100
    df_join['EW2_ALBEDO']=(df_join.EW2_SW_out/df_join.EW2_SW_in)*100
    df_join['EW3_ALBEDO']=(df_join.EW3_SW_out/df_join.EW3_SW_in)*100
    df_join.EW1_ALBEDO[df_join.EW1_ALBEDO<0]=0
    df_join.EW2_ALBEDO[df_join.EW2_ALBEDO<0]=0
    df_join.EW3_ALBEDO[df_join.EW3_ALBEDO<0]=0
    df_join.EW1_ALBEDO[df_join.EW1_ALBEDO>100]=np.NaN
    df_join.EW2_ALBEDO[df_join.EW2_ALBEDO>100]=np.NaN
    df_join.EW3_ALBEDO[df_join.EW3_ALBEDO>100]=np.NaN
    
    return df_join
```

```python
def join_ec_wrf_data():
    ec_stat=join_ec_data()
    print 'Read EC stat'
    dfw1 = pd.read_table('/media/klein-c/Elements/sens_stud/files/lu_grp/ec_stat_w123.csv',delimiter=',', parse_dates=['DATE'])
    dfw1.set_index('DATE', inplace=True, drop=True)
    df_join=ec_stat.join([dfw1])
    
    df_join.to_csv('/media/klein-c/Elements/sens_stud/files/lu_grp/ec_wrf_merge_swin.csv')
    
    aug=df_join.loc['20130801':'20130901']
    
    aug.to_csv('/media/klein-c/Elements/sens_stud/files/lu_grp/ec_wrf_merge_aug_swin.csv')
    
    return 'Finished'
```
